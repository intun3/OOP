## Первичный анализ данных

### Общее описание набора данных Blood Transfusion Service Center

Данный набор данных содержит информацию о донорстве крови в Центре переливания крови. Он включает 748 экземпляров с 5 признаками, из которых 4 являются численными, а 1 - бинарным целевым атрибутом.

### Описание признаков

1. **Recency** (Давность последнего пожертвования) - численный атрибут
2. **Frequency** (Частота пожертвований) - численный атрибут
3. **Monetary** (Денежная сумма пожертвований) - численный атрибут 
4. **Time** (Время с последнего пожертвования) - численный атрибут
5. **Target** (Сдача крови в текущем периоде) - бинарный целевой атрибут

### Характеристики данных

- Количество элементов (Instances): 748
- Количество признаков (Features): 5
- Связанные задачи: Бинарная классификация
- Характеристики набора данных: Многомерные (Multivariate)
- Тип признаков: Численные (Numeric), Бинарный (Binary)

### Распределение целевой переменной "Target"

Распределение целевой переменной "Target" следующее:
- 0 (Не сдавал кровь в текущем периоде): 570
- 1 (Сдавал кровь в текущем периоде): 178

Это показывает, что большинство участников (около 76%) не сдавало кровь в текущем периоде, что важно учитывать при построении моделей бинарной классификации.

Таким образом, данный набор данных Blood Transfusion Service Center представляет интерес для разработки и оценки моделей прогнозирования донорства крови на основе предоставленных характеристик доноров.

## Задание

Импортируем необходимые библиотеки

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

Загружаем данные из CSV-файла в Pandas DataFrame.
Выводим первых 5 строк данных.

```python
data = pd.read_csv('blood_transfusion.csv')
data.head()
```

Выводим информацию о DataFrame data

```python
data.info()
```

Создаём гистограмму распределения значений в столбце "Monetary (c.c. blood)" DataFrame data.

```python
data['Monetary (c.c. blood)'].hist()
```

Создаём столбчатую диаграмму, где:

ось X - индексы строк в DataFrame data
ось Y - значения столбца "Monetary (c.c. blood)" в DataFrame data

```python
plt.bar(data.index, data['Monetary (c.c. blood)'])
plt.show()
```

Подсчитываем количество уникальных значений в столбце "Frequency (times)" DataFrame data и сохраняем результат в переменную hist

Создаём столбчатую диаграмму, где:

ось X - уникальные значения из столбца "Frequency (times)"
ось Y - количество (частота) появления каждого значения

```python
hist = data['Frequency (times)'].value_counts()
plt.bar(hist.index, hist)
```

Определяем список столбцов (признаков), которые будут использованы для визуализации.
Создаём матрицу парных графиков (pairplot) с помощью библиотеки Seaborn, где:

Входные данные - DataFrame data, содержащий столбцы из списка feats и дополнительно столбец 'whether he/she donated blood in March 2007'.
Параметр hue='whether he/she donated blood in March 2007' - отображает значения этого столбца в виде разных цветов на графиках.

```python
feats = ['Recency (months)', 'Frequency (times)', 'Monetary (c.c. blood)', 'Time (months)']
sns.pairplot(data[feats + ['whether he/she donated blood in March 2007']], hue='whether he/she donated blood in March 2007')
```

Создаём список цветов для точек на графике, где "синий" обозначает "не сдавал(а) кровь", а "красный". - "сдавал(а) кровь".
Создаём список цветов для границ точек

Создаём scatter-plot, где:
ось X - значения столбца "Monetary (c.c. blood)"
ось Y - значения столбца "Time (months)"
цвет точек и их границ определяется списками c и edge_c

```python
c = data['whether he/she donated blood in March 2007'].map({0: 'lightblue', 1: 'orange'})
edge_c = data['whether he/she donated blood in March 2007'].map({0: 'blue', 1: 'red'})
plt.scatter(data['Monetary (c.c. blood)'], data['Time (months)'], color=c, edgecolors=edge_c)
plt.xlabel('Monetary (c.c. blood)')
plt.ylabel('Time (months)')
```
